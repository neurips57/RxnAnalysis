```python
import numpy as np
from scipy.stats import skew
import pandas as pd
```

```python
df = pd.read_csv('../Dataset/BH/bh_01.csv')
```

```python
df['Output'] = df['Output'].astype(float)
data1 = df.Output
```

```python
# Assuming you have data1 and data2 arrays as described
data1_sampled = []
np.random.seed(12)
# Set the desired total number of samples
total_samples = 200

# Define the weights function to control the mean value
mean_target = 10
range_min = 0
range_max = 100
# Calculate the weights based on distance from the mean and proximity to the desired range
weights = np.power(np.abs(data1 - mean_target), -4) * np.power(data1, 2)

#print(weights)
# Normalize the weights to ensure they sum up to 1
weights /= np.sum(weights)
#print(weights)
# Perform sampling using the probability distribution
sampled_indices = np.random.choice(len(data1), size=total_samples, replace=False, p=weights)
data1_sampled = data1[sampled_indices]

# Print the mean value of the sampled data
mean_sampled = np.mean(data1_sampled)
print("Mean of Sampled Data: ", mean_sampled)

# Print the skewness of the sampled data
skewness_sampled = skew(data1_sampled)
print("Skewness of Sampled Data: ", skewness_sampled)

# Print the median value of the sampled data
median_sampled = np.median(data1_sampled)
print("Median of Sampled Data: ", median_sampled)
```

```python
import matplotlib.pyplot as plt
# Plot the data distribution after sampling
plt.figure(figsize=(10, 5))
plt.hist(data1_sampled, bins=30, density=True, alpha=0.5, label='Sampled Data')
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Data Distribution - After Sampling')
plt.legend()
plt.show()
```
